12. Зависит ли **стоимость** игрока (Value) от его "**финтов**" (Skill Moves)?

```python
def money(x):
  x = x[1:]
  if x[-1] =='K':
    return float(x[:-1])*1000
  elif x[-1] =='M':
    return float(x[:-1])*1000000
  elif x != '':
    return float(x)
  else:
    return 0
df['Wage1'] = df['Wage'].apply(money)
df['Value1'] = df['Value'].apply(money)
```

```python
for i in range(1, 6):
  x = df.loc[df['Skill Moves'] == i, 'Value1']
  plt.title("Распределения стоимости игроков с международным рейтингом " + str(i))
  plt.hist(x, bins = np.arange(0, 120000000, 5000000))
  plt.show()
```

приведенные выше гистограммы показывают распрделение ценностей игроков с количеством финтов 1-5. Видно, что при увеличении количества финтов в среднем стоимость возрастает.

13. Покажите распределения основных характеристик игроков: **рост**, **вес**, **возраст** по **позициям** (Position), сделайте выводы

```python
def weight(x):
  if type(x) == float:
    return x
  else:
    return int(x[:-3])
```

```python
def height(x):
  if type(x) ==float:
    return x
  return 30.48 * float(x[0]) + 2.54*float(x[2:]) 
```

```python
df['Height1'] = df['Height'].apply(height)
df['Weight1'] = df['Weight'].apply(weight)
```

```python
pos = ['LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW', 'LAM', 'CAM', 'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM', 'CDM', 'RDM', 'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB']
pos1 = np.array(pos)
ph = np.zeros(len(pos))
pw = np.zeros(len(pos))
pa = np.zeros(len(pos))
for i in range(len(pos)):
  h = df.loc[df['Position'] == pos[i], 'Height1']
  w = df.loc[df['Position'] == pos[i], 'Weight1']
  a = df.loc[df['Position'] == pos[i], 'Age']
  ph[i] = sum(h)/len(h)
  pw[i] = sum(w)/len(w)
  pa[i] = sum(a)/len(a)
plt.figure(figsize=(10, 10))
plt.title("Зависимость средней высоты от позиции")
plt.scatter(pos1, ph)
plt.show()
plt.figure(figsize=(10, 10))
plt.title("Зависимость среднего веса от позиции")
plt.scatter(pos1, pw)
plt.show()
plt.figure(figsize=(10, 10))
plt.title("Зависимость среднего возраста от позиции")
plt.scatter(pos1, pa)
plt.show()
```

Наибольший рост в среднем у LCB, CB, LRB - у защитников

Наименьший рост в среднем у LF - левых форвардов

Найбольший вес в среднем у LCB, RCB - защитников

Наименьший вес в среднем у RAM - полузащитников

Найбольший возраст в среднем у LAM, CAM - полузащитников

Наименьший возраст в среднем у LW, CF, RW, CM - полузащитников и защитников

Все средние параметры отличаются друг от друга не сильно

26. Насколько средняя зарплата (Wage) игроков с рейтингом (Overall) 90+ больше средней зарплаты с рейтингом [80 - 90]? Визуализируйте возможную зависимость

```python
wage90 = df.loc[df['Overall'] > 90, 'Wage1']
wage80 = df.loc[(df['Overall'] >= 80) & (df['Overall'] <= 90), 'Wage1']
av90 = sum(wage90) / len(wage90)
av80 = sum(wage80) / len(wage80)
print(av80, av90)
v1 = []
w1 = []
for i in range(100):
  wagei = df.loc[df['Overall'] == i, 'Wage1']
  if len(wagei != 0):
    w1.append(sum(wagei) / len(wagei))
    v1.append(i)
v = np.array(v1)
w = np.array(w1)
plt.title("Зависимость средней стоимости игрока от рейтинга")
plt.plot(v, w)
plt.show()
```

средняя зарплата игроков с рейтингом >90: €386K

средняя зарплата игроков с рейтингом 80-90: €84K

Разница: €302K
